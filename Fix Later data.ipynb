```python
# Modules
import os
import cv2
import numpy as np
```

# Cropping Selected Area by the Mouse 

```python
def mouse_crop(event, x, y, flags, param):
    # grab references to the global variables
    global x_start, y_start, x_end, y_end, cropping, refPoint
 
    # if the left mouse button was DOWN, start RECORDING
    # (x, y) coordinates and indicate that cropping is being
    if event == cv2.EVENT_LBUTTONDOWN:
        x_start, y_start, x_end, y_end = x, y, x, y
        cropping = True
 
    # Mouse is Moving
    elif event == cv2.EVENT_MOUSEMOVE:
        if cropping == True:
            x_end, y_end = x, y
 
    # if the left mouse button was released
    elif event == cv2.EVENT_LBUTTONUP:
        # record the ending (x, y) coordinates
        x_end, y_end = x, y
        cropping = False # cropping is finished
 
        refPoint = [(x_start, y_start), (x_end, y_end)]
```

# Fix Collage images
- The code shows a window with the collage image
- We then select area of the image using the mouse
 - `O`: If we click `o` on the keyboard it means the selected area contains real person and it's moved to the original folder
 - `F`: If we click `f` it means the selected area contains a fake image generated by **faceapp**
 - `Q`: If we click `q` it will take us to the next image
 
## Illustration of the output

![](imgs/filterdata.png)

```python
# get the files names
files = os.listdir('later')

for file in files:
    # prepare the image window and callback
    cv2.namedWindow("image")
    cv2.setMouseCallback("image", mouse_crop)
    
    cropping = False
 
    x_start, y_start, x_end, y_end = 0, 0, 0, 0
    refPoint = []

    image = cv2.imread(f"later/{file}")
    oriImage = image.copy()
    
    while True:

        i = image.copy()

        #if not cropping:
        cv2.imshow("image", image)

        #elif cropping:
        cv2.rectangle(i, (x_start, y_start), (x_end, y_end), (255, 0, 0), 2)
        cv2.imshow("image", i)

        key = cv2.waitKey(1) & 0xFF


        if key == ord('o'):
            if len(refPoint) == 2: #when two points were found
                roi = oriImage[refPoint[0][1]:refPoint[1][1], refPoint[0][0]:refPoint[1][0]]
                cv2.imshow("Original", roi)
                # save cropped image to Original folder
                cv2.imwrite(f"original/{file.split('.')[0]}_original.png", roi)

        if key == ord('f'):
            if len(refPoint) == 2: #when two points were found
                roi = oriImage[refPoint[0][1]:refPoint[1][1], refPoint[0][0]:refPoint[1][0]]
                cv2.imshow("Fake", roi)
                # save cropped image to Original folder
                cv2.imwrite(f"fake/{file.split('.')[0]}_fake.png", roi)


        if key == ord('q'):
            break


    # close all open windows
    cv2.destroyAllWindows()
```
